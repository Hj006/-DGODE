```python
from imagebind.models import imagebind_model
from imagebind import data
import torch

device = "cuda" if torch.cuda.is_available() else "cpu"

# 加载模型并放到 GPU
model = imagebind_model.imagebind_huge(pretrained=True)
model = model.to(device)
model.eval()
print(f" ImageBind model loaded successfully on {device}!")

# 文本输入
text = ["a sound of a dog barking"]

# 加载并转换文本特征，放到 GPU
inputs = {imagebind_model.ModalityType.TEXT: data.load_and_transform_text(text, device)}

# 前向计算
with torch.no_grad():
    embeddings = model(inputs)

# 输出形状
print(" Embedding shape:", embeddings[imagebind_model.ModalityType.TEXT].shape)
```
